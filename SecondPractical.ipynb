**1. PLOTTING HISTOGRAMS USING PLOTLY**

```python
import plotly.express as px

fig = px.histogram(
    dataset,
    x="Income",
    title="Histogram of Citizens Incomes",
    width=600,
    height=400
)
fig.show()
```

**2. PLOTTING BAR CHARTS USING PLOTLY**

```python
import plotly.express as px

fig = px.bar(
    dataset,
    x="Education",
    y= "Income",
    title="Charts of Citizens' Incomes vs Education",
    width=600,
    height=400
)
fig.show()
```

**3. PLOTTING BOX PLOTS CHARTS USING PLOTLY**

```python
import plotly.express as px

fig = px.box(
    dataset,
    x="Education",
    y= "Income",
    title="Charts of Citizens Incomes vs Education",
    width=600,
    height=400,
    color='Education'
)
fig.show()
fig.write_html("my_box_chart.html")
```

**9. ADDING DROP DOWN MENU TO VISUALIZATIONS**

```python
import plotly.graph_objects as go
import numpy as np


plot = go.Figure(data=[go.Scatter(
    x=dataset['Education'],
    y=dataset['Income'],
    mode='markers',)
])

# Add dropdown
plot.update_layout(
    updatemenus=[
        dict(buttons=list([
            dict(
                args=["type", "scatter"],
                label="Scatter Plot",
                method="restyle"
            ),
            dict(
                args=["type", "bar"],
                label="Bar Chart",
                method="restyle"
            ),
            dict(
                args=["type", "histogram"],
                label="Histogram",
                method="restyle"
            ),
            dict(
                args=["type", "box"],
                label="Box Chart",
                method="restyle"
            )
        ]),
            direction="down",
        ),
    ]
)

plot.show()

```

**10. ADDING DROP DOWN MENU TO VISUAlIZATIONS**

```python

plot = go.Figure(data=[go.Scatter(
	x=dataset['Education'],
	y=dataset['Income'],
	mode='markers',  )
])

# Add dropdown
plot.update_layout(
	updatemenus=[
		dict(
			type="buttons",
			direction="left",
			buttons=list([
				dict(
					args=["type", "scatter"],
					label="Scatter Plot",
					method="restyle"
				),
				dict(
					args=["type", "bar"],
					label="Bar Chart",
					method="restyle"
				)
			]),
		),
	]
)

plot.show()
```

**11. ADDING SLIDERS TO THE PLOTS**

```python
x = dataset['Age']
y = dataset['Income']

plot = go.Figure(data=[go.Scatter(
	x=x,
	y=y,
	mode='markers',)
])

# Remove the rangeselector
plot.update_layout(
    xaxis=dict(
        rangeslider=dict(
            visible=True
        ),
    )
)

plot.show()
```

```python
x = dataset['Age']
y = dataset['Income']

plot = go.Figure(data=[go.Scatter(
	x=x,
	y=y,
	mode='markers',)
])

plot.update_layout(
	xaxis=dict(
		rangeselector=dict(
			buttons=list([
				dict(count=1,
					step="year",
					stepmode="backward",
                    label="1 Year Back"), # Label for the button
			])
		),
		rangeslider=dict(
			visible=True
		),
        title="Age" # Label for the x-axis
	),
    yaxis_title="Income" # Label for the y-axis
)

plot.show()
```
